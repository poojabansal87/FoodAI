```python
import pandas as pd 
```

```python
df = pd.read_csv("allergen_class.csv", encoding="iso-8859-1")
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df = df.dropna(how='any')
```

```python
df['AllergenClass'].nunique()
```

```python
new = df['AllergenClass'].str.split(',', expand=True) 
     
```

```python
df['MainAllergen']=new[0]
```

```python
df = df.loc[df['MainAllergen']!="None"]
```

```python
df = df.reset_index()
```

```python
for allergen in df['MainAllergen']:
    allergen = allergen.lower()
    df['main_allergen'] = allergen
```

```python
df.to_csv('cleaned_allergen_class.csv')
```
